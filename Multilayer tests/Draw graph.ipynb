```python
def drawGraph(fileName):
  plt.clf()
  outfile = fileName + '.png'
  A = pd.read_csv(fileName, sep=" ", header=None)
  G = nx.from_numpy_matrix(np.array(A))
  labels = dict()
  for row in A:
    labels[row] = row

  pos = nx.kamada_kawai_layout(G)
  G = nx.relabel_nodes(G, labels)
  nx.draw_networkx(G, with_labels = True, pos=pos, node_size = 20, font_size=3, width=0.2)
#   Check with this draw function as well, last graph looks weird
#   nx.draw_networkx(G, with_labels = True)
  print(outfile)
  plt.savefig(outfile, dpi=300)
  return G
```

```python
for fileName in test2:
  drawGraph(fileName)
```

```python
fileNames = ['aarhus_1.txt', 'aarhus_2.txt', 'aarhus_3.txt', 'aarhus_4.txt', 'aarhus_5.txt']
for fileName in fileNames:
  drawGraph(fileName)
print('Done')
```
